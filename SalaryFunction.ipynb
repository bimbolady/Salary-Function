```python
#A Python function that accepts employee names and returns details.
#Creation of employee function
#Function getEmployeeDetails receives a name as an argument. The name should be a string
"""
If the name exists inside the imported csv file, the employee detail is returned as a 
dataframe which is a peculiar data structure from the pandas package
"""

def getEmployeeDetails(name = None):
    if(name == None):
        print("Please enter a valid name in order for us to appropriately search")
    else:
       from IPython.display import display
       import pandas as pd
       # Error handling in your Python code.
       try:
         df = pd.read_csv("./Total.csv/Total.csv", low_memory=False)
         result_upper = df.query("EmployeeName == '{0}'".format(name.upper()))
         result_lower = df.query("EmployeeName == '{0}'".format(name.lower()))
         result_unchanged = df.query("EmployeeName == '{0}'".format(name))
         if len(result_upper) == 0 and len(result_lower) == 0 and len(result_unchanged) == 0:
            print("There is no data for this name")
            return
         elif len(result_upper) > 0:
             result_upper.to_csv("./Employee Profile/{0}.csv".format(name.upper()))
             return result_upper
         elif len(result_lower) > 0:
             result_lower.to_csv("./Employee Profile/{0}.csv".format(name.lower()))
             return result_lower  
         else:
             result_unchanged.to_csv("./Employee Profile/{0}.csv".format(name))
             return result_unchanged
       #if the csv file is not found, a FileNotFoundError exception is thrown but this is gracefully handled by our code.
       except FileNotFoundError:
         print("File not found! Please confirm that the designated file path is correct.")


#The employee details function in action.
display(getEmployeeDetails("Daniel W Plautz"))

```

```python
#Export Employee Details:
#This function initializes a zip folder
def exportEmployeeDetail(name):
    from IPython.display import display
    import pandas as pd
    import shutil

    if(name == None):
        print("Please enter a valid name in order for us to appropriately search")
    else:
       from IPython.display import display
       import pandas as pd
       import os
       #create directory for employee details
       pathToCreate = "C:\\Users\\Oremoxie\\data-analysis-assignment\\Employee Profile"
       try:
           if not os.path.exists(pathToCreate):
               os.makedirs(pathToCreate)
       except OSError:
           print("There was an issue creating the folder")
           
       # Error handling in your Python code.
       
       try:
         df = pd.read_csv("./Total.csv/Total.csv", low_memory=False)
         result_upper = df.query("EmployeeName == '{0}'".format(name.upper()))
         result_lower = df.query("EmployeeName == '{0}'".format(name.lower()))
         result_unchanged = df.query("EmployeeName == '{0}'".format(name))
         if len(result_upper) == 0 and len(result_lower) == 0 and len(result_unchanged) == 0:
            print("There is no data for this name")
            return
         elif len(result_upper) > 0:
             src = "./{0}.csv".format(name.upper())
             result_upper.to_csv(src)
             shutil.copy2(src, pathToCreate)
             shutil.make_archive("Employee Profile.zip", "zip", pathToCreate)
             # shutil.
             return result_upper
         elif len(result_lower) > 0:
             src = "./{0}.csv".format(name.lower())
             result_lower.to_csv(src)
             shutil.copy2(src, pathToCreate)
             shutil.make_archive("Employee Profile.zip", "zip", pathToCreate)
             return result_lower  
         else:
             src = "./{0}.csv".format(name)
             result_unchanged.to_csv(src)
             shutil.copy2(src, pathToCreate)
             shutil.make_archive("Employee Profile.zip", "zip", pathToCreate)
             return result_unchanged
       #if the csv file is not found, a FileNotFoundError exception is thrown but this is gracefully handled by our code.
       except FileNotFoundError:
         print("File not found! Please confirm that the designated file path is correct.")

#function call example
exportEmployeeDetail('SARAH CARY')
    
```
